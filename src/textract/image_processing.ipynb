```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "./images/sample1.jpg"

image = cv2.imread(file)

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
kernel = np.array([[0, -1, 0],
[-1, 5, -1],
[0, -1, 0]])
filtered_image = cv2.medianBlur(image, 1)

sharpened = cv2.filter2D(filtered_image, -1, kernel)
```

```python
print("Before sharpening:")
plt.imshow(image)
plt.show()
print("After sharpening:")
plt.imshow(sharpened)
plt.show()
```
